```python
import csv
def create_adjacency_list(csv_filename):
    adjacency_list = {}
    with open(csv_filename, 'r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            node1, node2 = row['node1'], row['node2']
            route_distance, heuristic_distance = int(row['route_distance']), int(row['Heuristic distance'])
            
            # Add an edge from node1 to node2 with the route distance
            if node1 in adjacency_list:
                adjacency_list[node1].append((node2, heuristic_distance))
            else:
                adjacency_list[node1] = [(node2, heuristic_distance)]

    return adjacency_list

# Example usage
csv_filename = 'CityGraph/random_connected_graph.csv'  # Replace with your CSV file name
adjacency_list = create_adjacency_list(csv_filename)

# Print the adjacency list
# for node, neighbors in adjacency_list.items():
#     print(f"{node}: {neighbors}")
print(adjacency_list)
```
