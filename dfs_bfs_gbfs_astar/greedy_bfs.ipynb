```python
# heulistic_distance up to target_node.

def heuristic_distance(target_node):
    heuristic_dis = {}
    for i in adjacency_list_for_complete_graph:
        if i==target_node:
            for i, j in adjacency_list_for_complete_graph[i]:
                heuristic_dis[i] = j
    return heuristic_dis

# print(heuristic_distance('Agra'))
# print(type(heuristic_distance('Agra')))
```
